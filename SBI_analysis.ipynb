```python
from Inference import (
    create_be_simulator,
    create_prior,
    train_sbi,
    run_all_diagnostics
)
from Helpers.utils import generate_stimuli
```

```python
# Setup
stimuli, categories, _ = generate_stimuli(n_trials=300, seed=42)
simulator = create_be_simulator(stimuli, categories, burn_in=100)
prior = create_prior()

```

```python
true_params = {
    'sigma_percep': 0.15,    # Perceptual noise
    'A_repulsion': 0.1,      # Serial dependence (repulsion)
    'eta_learning': 0.35,    # Learning rate (PRIMARY MEASURE for your research)
    'eta_relax': 0.12        # Relaxation rate toward uniform
}
true_params = [true_params[key] for key in ['sigma_percep', 'A_repulsion', 'eta_learning', 'eta_relax']]
```

```python
# Get observed data summary stats
observed_stats = simulator(true_params, seed=123)

# Train
result = train_sbi(
    simulator=simulator,
    prior=prior,
    observed_stats=observed_stats,
    method='NPE',  # or 'NLE', 'NRE'
    n_simulations=50000
)

# Sample posterior
samples = result.sample(10000)
print(result.summary())

# Validate
diagnostics = run_all_diagnostics(
    simulator, prior, result.posterior, observed_stats
)
```
